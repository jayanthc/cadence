# Sample analysis 

```python
from scipy.io import wavfile
import IPython.display as ipd
import numpy as np
import scipy.signal as ss
import matplotlib.pyplot as plt
```

```python
def unpack_file(fname):
    ipd.Audio(fname)
    rate, data = wavfile.read(fname)
    print('Sampling (frame) rate = {}'.format(rate))
    print('Total samples (frames) = {}'.format(data.shape[0]))
    duration = data.shape[0] / rate
    print('Duration = {:.1f} s'.format(duration))
    return rate, data, duration
```

```python
fname = 'data/wav/sample2.wav'
rate, data, duration = unpack_file(fname)
```

```python
def plot_time_signal(data):
    plt.figure(figsize=(20, 5))
    plt.plot(data)
    ticks, labels = plt.xticks()
    plt.xticks(ticks, map(lambda val: '{:.0f}'.format(val), ticks / rate))
    plt.xlim(0, duration * rate)
    plt.show()
```

```python
plot_time_signal(data)
```

```python
def convert_to_float(data):
    data = np.array([float(x) for x in data])
    print(data.dtype)
    return data 

def normalize(data):
    data_mean = np.mean(data)
    data -= data_mean
    data = np.abs(data)
    return data 
```

```python
data = convert_to_float(data)
data = normalize(data)
```

```python
def plot_samples(data):
    plt.figure(figsize=(20, 5))
    plt.plot(data)
    plt.show()
```

```python
def smooth_plot(data):
    out = data
    window = np.ones((10000,))
    out = np.convolve(data, window)
    out = (out - out.mean()) / out.std()
    plt.figure(figsize=(10, 5))
    plt.plot(out)
    plt.show()
    return out 
```

```python
plot_samples(data)
```

```python
def plot_fft(smooth_data):
    nfft = len(smooth_data)
    print(nfft)
    data_ft = np.abs(np.fft.rfft(smooth_data, nfft))
    data_ft[0] = 0
    
    tsamp = 1.0 / rate
    print(tsamp)
    data_freq = np.fft.rfftfreq(nfft, d=tsamp)

    upper_limit = 5    # Hz (corresponding to 300 bpm)
    delta_f = (rate / 2) / len(data_ft)
    len_limit = int(np.round(upper_limit / delta_f))
    
    plt.figure(figsize=(20, 5))
    plt.plot(data_freq[:len_limit], data_ft[:len_limit])
    plt.show()
```

```python
plot_fft(data)
```
